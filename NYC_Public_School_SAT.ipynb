```python
import pandas as pd

# Read in the data
schools = pd.read_csv('~/Library/Mobile Documents/com~apple~CloudDocs/LC_DA/Git/NYC_School_Test_Project/schools.csv')

# Preview the data
schools.head()
```

```python
schools['math_percent']=(schools['average_math']/schools['average_math'].max())*100
schools.head()
```

```python
eighty=schools[schools.math_percent >= 80]
eighty.head()
best_math_schools = eighty[['school_name', 'average_math']].sort_values('average_math', ascending=False)
best_math_schools.head()
```

```python
schools['total_SAT'] = schools['average_math']+schools['average_reading']+schools['average_writing']
schools.head()
```

```python
top_10_schools = schools.sort_values('total_SAT', ascending=False).iloc[:11]
top_10_schools = top_10_schools[['school_name', 'total_SAT']]
top_10_schools
```

```python
largest_std_dev = schools.groupby('borough')['total_SAT'].agg(['count', 'mean', 'std']).round(2)
largest_std_dev.head()
```

```python
new_columns={'count':'num_schools', 'mean':'average_SAT', 'std':'std_SAT'}
largest_std_dev.rename(columns=new_columns, inplace=True)
largest_std_dev.head()
```
